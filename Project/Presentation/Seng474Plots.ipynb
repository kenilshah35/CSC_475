## Helper function


```python
import matplotlib.pyplot as plt
import numpy as np

def plot_error(classifiers, errors, title):
    # plot the train and test errors of classifiers in a bar graph
    ax = plt.subplot(111)
    _X = np.arange(len(classifiers))
    colors = [['royalblue', 'darkmagenta', 'gold', 'green', 'tomato'], 
              ['cornflowerblue', 'orchid', 'yellow', 'lightgreen', 'lightsalmon']]
    for i in range(2):
        ax.bar(_X - 0.25 + i/4.0, errors[i], width=0.25, align='edge', color=[colors[i][j] for j in _X])
    plt.xticks(_X, classifiers)
    ax.set_ylabel('error')
    ax.set_ylim([0.0, 0.5])
    ax.legend(['train error', 'test error'])
    ax.set_title(title)

classifiers = ['Decision Trees', 'Random Forests', 'K nearest neighbors', 'Logistic Regression', 'SVM']
```

## Performance of Classifiers without Feature Selection

```python
n_train_error = [0.3326612903225806, 0.2943548387096774, 0.41491935483870956, 0.31612903225806455, 0.21451612903225814]
n_test_error = [0.3451612903225807, 0.3016129032258065, 0.4419354838709677, 0.2983870967741935, 0.1919354838709677]
n_errors = [n_train_error, n_test_error]
plt.figure(figsize = (10,6), dpi=300)
plot_error(classifiers, n_errors, 'Performance without Feature Elimination')
plt.show()
```

# Performance of Classifiers with Tree Based Feature Selection

```python
t_train_error = [0.35806451612903223, 0.282258064516129, 0.4193548387096774, 0.3217741935483871, 0.2338709677419355]
t_test_error = [0.3709677419354839, 0.27903225806451615, 0.4419354838709677, 0.30483870967741933, 0.2129032258064516]
t_errors = [t_train_error, t_test_error]
plt.figure(figsize = (10,6), dpi=300)
plot_error(classifiers, t_errors, 'Performance with Tree Based Feature Elimination')
plt.show()
```

## Performance of Classifiers with Recursive Feature Elimination

```python
r_train_error = [0.2887096774193548, 0.21935483870967742, 0.32096774193548394, 0.3189516129032258, 0.21572580645161288]
r_test_error = [0.26290322580645165, 0.20322580645161292, 0.3129032258064516, 0.32258064516129037, 0.20483870967741935]
r_errors = [r_train_error, r_test_error]
plt.figure(figsize = (10,6), dpi=300)
plot_error(classifiers, r_errors, 'Performance with Recursive Feature Elimination')
plt.show()
```

## Performance with ANOVA F measure

```python
a_train_error = [0.3375, 0.28346774193548396, 0.3911290322580645, 0.3245967741935484, 0.22419354838709682]
a_test_error = [0.282258064516129, 0.2338709677419355, 0.3354838709677419, 0.2661290322580645, 0.15161290322580645]
a_errors = [a_train_error, a_test_error]
plt.figure(figsize = (10,6), dpi=300)
plot_error(classifiers, a_errors, 'Performance with ANOVA F measure')
plt.show()
```

## Performance Of All Classifiers with All Feature Selection

```python
x_labels =['None', 'Tree Based', 'Recursive Feature Elimination', 'ANOVA F measure']

dt_train_error = [n_train_error[0], t_train_error[0], r_train_error[0], a_train_error[0]]
dt_test_error = [n_test_error[0], t_test_error[0], r_test_error[0], a_test_error[0]]
rf_train_error = [n_train_error[1], t_train_error[1], r_train_error[1], a_train_error[1]]
rf_test_error = [n_test_error[1], t_test_error[1], r_test_error[1], a_test_error[1]]
knn_train_error = [n_train_error[2], t_train_error[2], r_train_error[2], a_train_error[2]]
knn_test_error = [n_test_error[2], t_test_error[2], r_test_error[2], a_test_error[2]]
lgr_train_error = [n_train_error[3], t_train_error[3], r_train_error[3], a_train_error[3]]
lgr_test_error = [n_test_error[3], t_test_error[3], r_test_error[3], a_test_error[3]]
svm_train_error = [n_train_error[4], t_train_error[4], r_train_error[4], a_train_error[4]]
svm_test_error = [n_test_error[4], t_test_error[4], r_test_error[4], a_test_error[4]]
train_errors = [dt_train_error, rf_train_error, knn_train_error, lgr_train_error, svm_train_error]
test_errors = [dt_test_error, rf_test_error, knn_test_error, lgr_test_error, svm_test_error]

plt.figure(figsize=(12,8))
test_colours = ['darkturquoise', 'darkmagenta', 'gold', 'green', 'tomato']
train_colours = ['turquoise', 'orchid', 'yellow', 'lightgreen', 'lightsalmon']
for i in range(len(classifiers)):
    plt.plot(x_labels, train_errors[i], label = "{0} train error".format(classifiers[i]), 
             color=train_colours[i], marker='o')
    plt.plot(x_labels, test_errors[i], label = "{0} test error".format(classifiers[i]), 
             color=test_colours[i], marker='o')
plt.title('Performance of Classifiers with Feature Selection')
plt.legend(bbox_to_anchor=(1.0, 1.0))
plt.show()
```
