```python

import matplotlib.pyplot as plt
import numpy as np
def plot_error(classifiers, accuracy, title):
    # plot the train and test errors of classifiers in a bar graph
    ax = plt.subplot(111)
    _X = np.arange(len(classifiers))
    colors = [['green', 'green', 'green', 'green','green'], 
              ['lightgreen', 'lightgreen', 'lightgreen', 'lightgreen', 'lightgreen']]
    for i in range(2):
        ax.bar(_X - 0.25 + i/4.0, accuracy[i], width=0.25, align='edge', color=[colors[i][j] for j in _X])
    plt.xticks(_X, classifiers)
    ax.set_ylabel('accuracy')
    #ax.set_ylim([0.0, 0.5])
    ax.legend(['train accuracy', 'test accuracy'])
    ax.set_title(title)
```

```python
classifiers = ['Baseline','TB','RFE','F-Test','Chi-2']
train = [0.342, 0.3857,0.6853,0.516,0.5171]
test = [0.322, 0.3559, 0.4775,0.644,0.28813]
accuracy = [train, test]
plt.figure(figsize = (10,6), dpi=300)
plot_error(classifiers, accuracy, 'Accuracy for Midi Features')
plt.savefig("Midi_features.png")
plt.show()
```

```python
classifiers = ['Base Midi', 'Base Lyrics', 'Midi+Lyrics']
train = [0.342, 0.3622 ,  0.42]
test = [0.322, 0.40 , 0.4646]
accuracy = [train, test]
plt.figure(figsize = (10,6), dpi=300)
plot_error(classifiers, accuracy, 'Accuracy for Midi+Lyrics Features')
plt.savefig("Combined_features.png")
plt.show()
```
