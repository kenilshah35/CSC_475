```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn import preprocessing
from sklearn.model_selection import cross_val_score
from sklearn import svm
from sklearn.feature_selection import RFECV
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import accuracy_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import chi2
from sklearn import metrics
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('audiofeaturestransformed.csv')
```

```python
df.head()
```

```python
cluster_mood_filename = df.cluster_mood_filename
temp = []
cluster = []
mood = []
filename = []

for i in cluster_mood_filename:
    temp.append(i)
temp = [i.split('\\') for i in temp]

for path in temp:
    cluster.append(path[-3])
    mood.append(path[-2])
    filename.append(path[-1].split('.')[0])

df['cluster'] = cluster
df['mood'] = mood
df['filename'] = filename

cols = [col for col in df.columns]
xcols = cols[1:-3]
ycol = cols[-3]
print(ycol)
```

```python
df.head()
```

```python
le = preprocessing.LabelEncoder()
X = df[xcols]
X = X.fillna(0)
Y = df[ycol]
Y = pd.DataFrame(Y)
Y = Y.apply(lambda col: le.fit_transform(col.astype(str)), axis=0, result_type='expand')
files = df['filename']
# print(type(df))
# print(type(X_train))
# print(type(Y_train))
# print(type(xcols))
# print(type(ycol))
# print(X.isnull().sum())
# print(Y.isnull().sum())
# print(X.shape)
# X_nonan = X.fillna(0)
# print(X_nonan.isnull().values.any())
# print(X_nonan.shape)
print(X.isnull().values.any())
print(X.shape)
print(Y.shape)
```

```python
normalized_X = X.values
std_scalor = preprocessing.MinMaxScaler()
X_scaled = std_scalor.fit_transform(normalized_X)
normalized_X = pd.DataFrame(X_scaled, columns=xcols)
#X_train3, X_test3, Y_train3, Y_test3 = train_test_split(normalized_X, Y, test_size=0.3)
X_train, X_test, Y_train, Y_test = train_test_split(normalized_X, Y, test_size=0.3, stratify = Y)
normalized_X.head()
```

```python
%%time
param = {'C': [0.001,0.01,0.1, 1, 10, 100, 1000],
              'kernel': ['linear']}
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=20)
grid = GridSearchCV(svm.SVC(probability=True), param, verbose = 3, cv=cv, return_train_score=True, n_jobs=15)
grid.fit(X_train, Y_train.values.ravel())
    
train_accuracy_svm_norm = grid.best_score_
    
grid_predict = grid.predict(X_test)
test_accuracy_svm_norm  = metrics.accuracy_score(Y_test,grid_predict)

print("Training accuracy " + str(train_accuracy_svm_norm))
print("Test accuracy " + str(test_accuracy_svm_norm))
print(grid.best_params_)
print(metrics.classification_report(Y_test, grid_predict))

grid_proba = grid.predict_proba(X_test)
auc = metrics.roc_auc_score(Y_test, grid_proba, multi_class="ovr")
print("Auc:", str(auc))
```

```python
%%time
clf = svm.SVC(C=10, kernel='linear')
#clf = LinearSVC(C=0.1, max_iter = 10000000)
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=20)
#cv = StratifiedKFold(5)
rfecv = RFECV(estimator = clf, step = 1, cv = cv, scoring='accuracy', n_jobs=15,min_features_to_select =5 )
rfetrain=rfecv.fit(X_train, Y_train.values.ravel())
print('Optimal number of features :', rfecv.n_features_)

rfe = RFE(estimator=clf, n_features_to_select=rfecv.n_features_, step=1)
rfe = rfe.fit(X_train, Y_train.values.ravel())

rfe_X_train = X_train.loc[:, rfe.get_support()]
rfe_X_test = X_test.loc[:, rfe.get_support()]

param = {'C': [0.01,0.1, 1, 10,100,1000,10000],
              'kernel': ['linear']}
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=20)
grid = GridSearchCV(svm.SVC(probability=True), param, verbose = 3, cv=cv, return_train_score=True, n_jobs=15)
grid.fit(rfe_X_train, Y_train.values.ravel())
    
train_accuracy_svm_rfe = grid.best_score_
    
grid_predict = grid.predict(rfe_X_test)
test_accuracy_svm_rfe  = metrics.accuracy_score(Y_test,grid_predict)

print("Training accuracy " + str(train_accuracy_svm_rfe))
print("Test accuracy " + str(test_accuracy_svm_rfe))
print(grid.best_params_)
print(metrics.classification_report(Y_test, grid_predict))

grid_proba = grid.predict_proba(rfe_X_test)
auc = metrics.roc_auc_score(Y_test, grid_proba, multi_class="ovr")
print("Auc:", str(auc))

for index, rank in enumerate(rfe.ranking_):
    if rank in [1]:
        print("Rank " + str(rank)+ " feature: " + xcols[index])
```
